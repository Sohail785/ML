```python
from selenium import webdriver
from bs4 import BeautifulSoup
import random
import os
import urllib
import urllib.request
```

```python
browser=webdriver.Chrome("C:/Users/Sohail/Desktop/chromedriver.exe")
browser.get('https://www.amazon.in/s?k=jeans+for+men&ref=nb_sb_noss_2')

```

```python
# Mens Jeans
# soup=BeautifulSoup(browser.page_source)
html = browser.page_source
soup = BeautifulSoup(html)


#do something useful
#prints all the links with corresponding text
all_images = []
for link in soup.find_all('div',{"data-component-type":"s-search-result"},limit=1000):
    href = link.img['src']
    print(href)
    
    img_name = random.randrange(1000,2000)
    
    full_name = str(img_name) + ".jpg"
    
    urllib.request.urlretrieve(href, full_name)
    
    print("loop break")
```
