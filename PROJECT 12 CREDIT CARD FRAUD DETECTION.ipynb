```python
x=data.drop("Class",axis=1)
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x=scale.fit_transform(x)
```

```python
y=data["Class"]
y.head()
```

```python
print(x.shape)
print(y.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.21,random_state=11)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lm=LogisticRegression()
lm.fit(x_train,y_train)
pred=lm.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
#ROC CURVE
from sklearn.metrics import roc_curve
```

```python
y_pred_prob=lm.predict_proba(x_test)[:,1]
y_pred_prob
```

```python
fpr,tpr,thresholds=roc_curve(y_test,y_pred_prob)
```

```python
print(fpr)
print(tpr)
print(thresholds)
```

```python
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label="LogisticRegression")
plt.title("Logistic Regression")
plt.show()
```

```python
from sklearn.metrics import roc_auc_score
```

```python
auc_score=roc_auc_score(y_test,lm.predict(x_test))
print(auc_score)
```

```python
#Applying Bagging Methods
from sklearn.ensemble import RandomForestClassifier
```

```python
rfr=RandomForestClassifier(n_estimators=300,random_state=42)
rfr.fit(x_train,y_train)
rfr.score(x_train,y_train)
```

```python

gbr=GradientBoostingClassifier()
gbr.fit(x_train,y_train)
gbr.score(x_train,y_train)
```
